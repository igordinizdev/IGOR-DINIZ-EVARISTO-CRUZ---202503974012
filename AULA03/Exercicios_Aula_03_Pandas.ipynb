```python
# ---- SETUP ----

import pandas as pd
import numpy as np
```

```python
# ---- CARREGAR O DATASET ----

df = pd.read_csv("livros.csv", sep=";")
print(f"Shape: {df.shape}")
```

```python
# ---- Nivel 1. EXPLORAÇÃO INICIAL ----

# 1 - Exibindo as 5 primeiras linhas
print(df.head())
```

```python
# 2 - Use info() e describe() para entender a estrutura. Anote: quais colunas têm tipo incorreto?

print(df.info())
print(df.describe)
```

```python
# 3 - Quantos valores nulos existem em cada coluna? Crie uma tabela resumo.

df.isnull().sum()
```

```python
# 4 - Existem livros com 0 páginas no dataset? Se sim, quantos? Filtre e exiba esses registros.

livros_zero_paginas = df[df["paginas"] == 0]

print("A quantidade de livros com 0 páginas é:", livros_zero_paginas.shape[0])

livros_zero_paginas
```

```python
# 5 - Quantos livros distintos foram publicados em cada ano? Use value_counts() e ordene cronologicamente.

df["ano"].value_counts().sort_index()
```

```python
# ---- Nivel 2. TRANSFORMAÇÃO E LIMPEZA ----

# 1 - Crie uma coluna faixa_paginas com as categorias: "Curto" (<150), "Médio" (150–350) e "Longo" (>350). Use apply() com lambda.

df["faixa_paginas"] = df["paginas"].apply(
    lambda x: "Curto" if x < 150 else ("Médio" if x <= 350 else "Longo")
)
```

```python
# 2 - Remova todos os registros onde paginas == 0 e salve em df_limpo. Quantos registros foram removidos?

df_limpo = df[df["paginas"] > 0].copy()

removidos = df.shape[0] - df_limpo.shape[0]
print("Foram removidos:", removidos, "registros")
```

```python
# 3 - Preencha os nulos da coluna ano com a mediana e converta para inteiro.

df_limpo["ano"] = df_limpo["ano"].fillna(df_limpo["ano"].median()).astype(int)

df_limpo["ano"].info()
```

```python
# 4 - Crie uma coluna decada representando a década de publicação (ex: 2019 → 2010, 1987 → 1980).

df_limpo["decada"] = (df_limpo["ano"] // 10) * 10

print(df_limpo)
```

```python
# ---- Nivel 3. ANÁLISE AVANÇADA (BONUS) ----

# 1 - Qual a média de páginas por década? Use groupby() e ordene da mais antiga para a mais recente.

media_paginas_por_decada = df_limpo.groupby("decada")["paginas"].mean().sort_index()

media_paginas_por_decada
```

```python
# 2 - Quais são os 10 autores com mais livros no dataset? Use value_counts() na coluna autor.

df_limpo["autor"].value_counts().head(10)
```

```python
# 3 - Qual a distribuição de faixa_paginas para livros publicados após 2010? Filtre e use value_counts().

df_limpo[df_limpo["ano"] > 2010]["faixa_paginas"].value_counts()
```

```python
# 4 - Exporte o df_limpo para um arquivo Excel chamado livros_analisados.xlsx.

df_limpo.to_excel("livros_analisados.xlsx", index=False)
```
